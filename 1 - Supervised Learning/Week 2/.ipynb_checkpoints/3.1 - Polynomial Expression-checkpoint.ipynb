## Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.linspace(0, 5, 100)
Y = (2 * (X**3)) + (3 * (X**2)) + (4 * X) - 5 + np.random.normal(0, 10, 100) # The Last Expression Is use For Scattering The Data Points

print(f"X = {X}")
print(f"Y = {Y}")
```

```python
# Normalizing
X_norm = (X - np.mean(X))/np.std(X)
Y_norm = (Y - np.mean(Y))/np.std(Y)

print(f"X Normalised = {X_norm}")
print(f"Y Normalised = {Y_norm}")
```

```python
# Ploting Data Points
plt.figure(figsize=(8, 6))
plt.scatter(X_norm, Y_norm, label="Original Data", color='blue')
plt.title("Polynomial Regression using Gradient Descent")
plt.xlabel("Size (Normalised) in feet sqr")
plt.ylabel("Price (Normalised) in $1000's")
plt.legend()
plt.grid(True)
```

## Convert To Polynomial Data Set

```python
def convert_to_polynomial(X, degree):
    X_poly = np.column_stack([X**i for i in range(degree + 1)])
    return X_poly
```

```python
test = np.array([2, 3, 4])
degree = 3
test_poly = convert_to_polynomial(test, degree)

print(f"Polynomial Test = {test_poly} With Degree = {degree}")
```

## Gradeint Descent Algorthim For Polynomial Regresssion

```python
def polynomial_gradient_descent(X, Y, W, b, alpha, degree, iterations):
    X_poly = convert_to_polynomial(X, degree)
    m, n = X_poly.shape
    W = np.zeros(n)
    
    for i in range(iterations):
        f_wb = np.dot(X_poly, W) + b
        
        error = f_wb - Y
        cost = np.sum(error ** 2) / (2*m)
        
        d_dw = (1/m) * np.dot(X_poly.T, error) # We can also update W Like This
        d_db = (1/m) * np.sum(error)
        
        W = W - alpha * d_dw
        b = b- alpha * d_db
        
        # Print the cost every few epochs
        if i % 100 == 0:
            print(f"Iteration {i}: Cost = {cost}")
    
    return W, b
```

```python
# Hyper Parameters
W = 0
b = 0
alpha = 0.01
degree = 3
iterations = 10000

W_updated, b_updated = polynomial_gradient_descent(X_norm, Y_norm, W, b, alpha, degree, iterations)
```

```python
print(f"W_updated = {W_updated}")
print(f"b_updated = {b_updated}")
```

## Prediction Model

```python
def predict(X, W, b):
    degree = W.shape[0] - 1 
    X_poly = convert_to_polynomial(X, degree)
    return np.dot(X_poly, W) + b
```

```python
Y_hat = predict(X_norm, W_updated, b_updated)
print(f"Predicted X Values (Y_hat) = {Y_hat}")
```

## Visualisation

```python
# Ploting Data Points
plt.figure(figsize=(8, 6))
plt.scatter(X_norm, Y_norm, label="Original Data", color='blue')
plt.title("Polynomial Regression using Gradient Descent")
plt.xlabel("Size (Normalised) in feet sqr")
plt.ylabel("Price (Normalised) in $1000's")
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_norm, Y_norm, label="Original Data", color='blue')
plt.plot(X_norm, Y_hat, label=f"Fitted Polynomial (Degree = {degree})", color='r')
plt.title("Polynomial Regression using Gradient Descent")
plt.xlabel("Size (Normalised) in feet sqr")
plt.ylabel("Price (Normalised) in $1000's")
plt.legend()
plt.grid(True)
```
