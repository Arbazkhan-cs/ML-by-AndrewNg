## First Lab For Simple Model Representation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])  # Price in $1000's
y_train = np.array([300, 500])  # Area in feet^2

print(f"X = {x_train}")
print(f"Y = {y_train}")
```

```python
m = len(x_train)

print(f"Total Number Of Training Example = {m}")
```

<pre>
-> As we have seen the function f (f = wx + b)
-> For one data point we have function:
        fw,b(x[0]) = wx[0] + b
-> So to find a model for supporting each point we have a loop with function
        fw,b(x[i]) = wx[i] + b
</pre>

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      y (ndarray (m,)): target values
    """
    
    m = len(x)
    f_wb = np.zeros(m)  # initail all are zeros
    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb
```

```python
# Let First take w = 100 and b = 100
w = 100
b = 100

print(f"w = {w}")
print(f"b = {b}")
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

plt.scatter(x_train, y_train, c='r', marker='x', label='Data Points')
plt.plot(x_train, tmp_f_wb, c='b', label="Prediction Line")
plt.legend()
plt.show()
```

So in above we can see that for the value of [w = 100 and b = 100] the predicted line is not good so let change the value of w = 200 and b = 100

```python
w = 200
b = 100

print(f"New Value Of w = {w}")
print(f"New Value Of b = {b}")
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

plt.scatter(x_train, y_train, c='r', marker='x', label="Data Point")
plt.plot(x_train, tmp_f_wb, c='b', label="Prediction Line")
plt.xlabel("Size 1000 sq_feet")
plt.ylabel("Price in $1000's")
plt.legend()
plt.show()
```
