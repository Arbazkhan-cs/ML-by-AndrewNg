## Softmax Activation Function

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Image

import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Sequential
from sklearn.datasets import load_digits
```

```python
Data = load_digits(n_class=5)
```

```python
X, Y = Data["data"], Data["target"]
X.shape, Y.shape
```

```python
X, Y
```

```python
def normalize(data):
    return (data-np.mean(data))/np.std(data)
```

```python
X_norm = normalize(X)
```

```python
X_norm
```

## Creating model with linear layer

```python
model = Sequential([
    Input(64),
    Dense(100, activation="relu", name="HiddenLayer1"),
    Dense(150, activation="relu", name="HiddenLayer2"),
    Dense(5, activation="linear")
])
```

```python
model.summary()
```

```python
model.compile(
    loss = tf.keras.losses.sparse_categorical_crossentropy,
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)
```

```python
model.fit(X_norm, Y, epochs=10)
```

```python
a = model.predict(X)
```

```python
a
```

## Adding Softmax Function

```python
# Specify the path to your image file
image_path = "./notes/Softmax function eqn.png"

# Display the image
display(Image(filename=image_path))
```

```python
def SoftMax(z):
    # classes = 5
    outputs = []
    for i in z:
        a = (np.exp(i))/np.sum(np.exp(z))
        outputs.append(a)
    
    print(outputs)
    return np.array(outputs)
```

```python
Predict = softmax(a)
```

## Prediction

```python
Predict[0]
```

```python
predict = [np.argmax(i) for i in Predict]
```
