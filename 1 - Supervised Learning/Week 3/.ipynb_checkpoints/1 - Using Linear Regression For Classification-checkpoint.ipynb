# Classification Problem With Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([5, 10, 15, 25, 30, 35])
y = np.array([0, 0, 0, 1, 1, 1])

print(f"x = {x}")
print(f"y = {y}")
```

```python
plt.scatter(x, y, color=["red" if label==0 else "blue" for label in y], label="Data Points")
plt.title("Classification")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
```

```python
def compute_cost(x, y, w, b):
    m = len(x)
    f_wb = np.dot(x, w) + b
    
    error = f_wb - y
    cost = np.sum(error**2)/(2*m)
    
    return cost
```

```python
def gradient_descent(x, y, w, b, alpha):
    m = len(x)
    iteration = 10000
    costT = []
    
    for i in range(iteration):
        f_wb = np.dot(x, w) + b
        
        cost = compute_cost(x, y, w, b)
        costT.append(cost)
        
        d_dw = (1/m)*np.sum((f_wb - y) * x)
        d_db = (1/m)*np.sum(f_wb - y)
        
        w = w - alpha * d_dw
        b = b - alpha * d_db
        
        if(i%100 == 0):
            print(f"Iteration = {i}, Cost = {cost}")
        
        if(costT[i] == costT[i-1] and i > 0):
            break
    
    return w, b
```

```python
w = 0
b = 0
alpha = 0.001

w_updated, b_updated = gradient_descent(x, y, w, b, alpha)
```

```python
def predict(x, w, b):
    return (np.dot(x, w) + b)
```

```python
f_wb = predict(x, w_updated, b_updated)
f_wb
```

```python
plt.scatter(x, y, color=["red" if label==0 else "blue" for label in y], label="Data Points")
plt.plot(x, f_wb, color="green", label="Regression Line")
# np.plot(20, [0.0, 0.2])
plt.title("Classification")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
```

```python
plt.scatter(x, y, color=["red" if label == 0 else "blue" for label in y], label="Data Points")
plt.plot(x, f_wb, color="green", label="Regression Line")

# Plot the decision boundary (for example, at threshold 0.5)
plt.axvline(np.average(x), color='purple', linestyle='--', label='Decision Boundary')

plt.title("Classification with Decision Boundary")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

For above data set our Linear Regression Model Is Working Model But What Happen if we increase the number of data set at y=1?

```python
x = np.array([5, 10, 15, 25, 30, 35, 65])
y = np.array([0, 0, 0, 1, 1, 1, 1])

w = 0
b = 0
alpha = 0.001

w_updated, b_updated = gradient_descent(x, y, w, b, alpha)
```

```python
f_wb = predict(x, w_updated, b_updated)
```

```python
plt.scatter(x, y, color=["red" if label == 0 else "blue" for label in y], label="Data Points")
plt.plot(x, f_wb, color="green", label="Regression Line")

# Plot the decision boundary (for example, at threshold 0.5)
plt.axvline(np.average(x), color='purple', linestyle='--', label='Decision Boundary')

plt.title("Classification with Decision Boundary")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

But Now we can clearly see that Linear Regression Model Is Worst In This scenerio
