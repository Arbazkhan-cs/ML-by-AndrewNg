### Gradient Descent Algorithm

```python
def compute_cost(x, y, w, b):
    m = len(x)
    
    f_wb = np.dot(x, w) + b
    
    cost = np.sum((f_wb - y)**2)
    cost = (1/(2*m)) * cost
    
    return cost

def gradient_descent(x, y, w, b, a):
    m = len(x)
    iteration = 10000
    total_cost = []
    
    for i in range(iteration):
        f_wb = np.dot(x, w) + b
        
        cost = compute_cost(x, y, w, b)
        print(f"Iteration = {i}, Cost = {cost}")
        
        d_dw = (1/m)*(np.sum((f_wb - y) * x))
        d_db = (1/m)*(np.sum(f_wb - y))
        
        w = w - a * d_dw
        b = b - a * d_db
        
        if(cost in total_cost):
            break
        
        total_cost.append(cost)
    
    return w, b
```

```python
w = 0
b = 0
a = 0.01
x_train = df3["scaler_Experience"]
y_train = df3["scaler_Salary"]

w_updated, b_updated = gradient_descent(x_train, y_train, w, b, a)
```

### Regression line

```python
def regression_line(x, w, b):
    m = len(x)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb

f_wb = regression_line(x_train, w_updated, b_updated)
```

```python
plt.scatter(df3["scaler_Experience"], df3["scaler_Salary"], marker='x', color='r', label="Data Point")
plt.plot(x_train, f_wb, color='b', label="Regression Line")
plt.title("Experience Vs Salary")
plt.ylabel("Experience")
plt.xlabel("Salary")
plt.legend()
plt.show()
```

### Prediction

```python
def predict(x, w, b):
    y_hat = x * w + b
    return y_hat
```
