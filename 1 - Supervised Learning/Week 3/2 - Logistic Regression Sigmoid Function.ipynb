# Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data with one feature
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([0, 0, 0, 1, 1, 1])

print(f"x = {x}")
print(f"y = {y}")
```

```python
plt.scatter(x, y, color=["red" if label==0 else "blue" for label in y], label="Data Points")
plt.title("Classification")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
```

```python
def sigmoid_fn(x, w, b):
    z = np.dot(x, w) + b
    return 1/(1 + np.exp(-z))
```

```python
w = 2.35392395
b = -8.180776624917412

x_values = np.linspace(min(x), max(x), 100)
gz = sigmoid_fn(x_values, w, b)
print(f"X_values = {x_values}")
print(f"gz = {gz}")
```

```python
plt.scatter(x, y, color=["red" if label==0 else "blue" for label in y], label="Data Points")
plt.plot(x_values, gz, color="green", label="Logistic Regression Line")

plt.title("Classification")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
```
