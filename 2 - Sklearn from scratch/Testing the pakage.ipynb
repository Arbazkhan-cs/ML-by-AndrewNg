# Importing The Our Sklearn Pakage

```python
import Sklearn as sk
from Sklearn import datasets
from Sklearn import model_selection

import numpy as np
import matplotlib.pyplot as plt
```

# Testing LinearRegression

### generating dataset

```python
X_test, Y_test = datasets.generate_linear_dataset()
```

```python
X_test.shape, Y_test.shape
```

### model building

```python
model = model_selection.LinearRegression(X_test, Y_test)
```

```python
model.fit()
```

```python
model.bestFitLine()
```

### Measuring performance with Predifined sklearn

```python
from sklearn.linear_model import LinearRegression
```

```python
model2 = LinearRegression()
```

```python
model2.fit(X_test, Y_test)
```

```python
y_hat = model.predict(X_test)
y_hat2 = model2.predict(X_test)
```

```python
plt.scatter(X_test[:, 0], Y_test, label="Data Points")
plt.plot(X_test[:, 0], y_hat2, color="red",  linewidth=4, label="sklearn Line")
plt.plot(X_test[:, 0], y_hat, color="orange", linewidth=1, label="my Sklearn Line")
plt.xlabel("Area")
plt.ylabel("price")
plt.title("Performance")
plt.legend()
plt.grid(True)
```

# Testing LogisticRegression

### generating dataset

```python
X_test, Y_test = datasets.generate_sigmoid_dataset(num_features=2)
```

```python
X_test.shape, Y_test.shape
```

### model building

```python
model = model_selection.LogisticRegression(X_test, Y_test)
```

```python
model.fit()
```

```python
model.plot_decision_boundary()
```

### Measuring performance with Predifined sklearn

```python
from sklearn.linear_model import LogisticRegression
```

```python
model2 = LogisticRegression()
```

```python
model2.fit(X_test, Y_test)
```

```python
y_hat = model.predict(X_test)
y_hat2 = model2.predict(X_test)
```

```python
# Plot data points
plt.scatter(X_test[:, 0], X_test[:, 1], c=Y_test, cmap=plt.cm.coolwarm, s=20, edgecolors='k')

# Plot decision boundary for my model
x1_min, x1_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
x2_min, x2_max = X_test[:, 1].min() - 1, X_test[:, 1].max() + 1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.1),
                       np.arange(x2_min, x2_max, 0.1))

Z = model.predict(np.c_[xx1.ravel(), xx2.ravel()])
Z = Z.reshape(xx1.shape)
plt.contour(xx1, xx2, Z, levels=[0.5], colors='orange', linestyles='--')

# Plot decision boundary for sklearn model
Z2 = model2.predict(np.c_[xx1.ravel(), xx2.ravel()])
Z2 = Z2.reshape(xx1.shape)
plt.contour(xx1, xx2, Z2, levels=[0.5], colors='blue', linestyles='--')

# Add labels
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Boundaries')
plt.show()
```
